```python
import pandas as pd # for dealing with data!
import matplotlib.pyplot as plt # for visualizing the data!
import numpy as np# for calculating the moving average!
```

```python
globaltemp = pd.read_csv('global_data.csv') # importing 'global tempreature data'
citytemp = pd.read_csv('city_data.csv') # importing 'city tempreature data'
```

```python
## Doing 'Moving Average' for the Global data and Local data.
```

```python
## Visualizng Data using matplotlib Library.
```

```python
globaltemp['avg_temp'].plot(label="Global Avg. Temp.", figsize=(15,8))
citytemp['avg_temp'].plot(label="Riyadh Avg. Temp.")
globaltemp['avg_temp'].rolling(window=10).mean().plot(label="10-Year MA Global Avg. Temp.")
citytemp['avg_temp'].rolling(window=10).mean().plot(label="10-Year MA Riyadh Avg. Temp.")
plt.legend(loc='center right', fancybox=True, framealpha=1, shadow=True, borderpad=1)
plt.grid(True)
plt.xlabel("10-Year MA")
plt.ylabel("Temperature (°C)") 
plt.title("Global VS. Local Average Temperature")
plt.show()
```
